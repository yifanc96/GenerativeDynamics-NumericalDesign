```python
import os
import numpy as np
import torch
import torch.nn as nn
import torch.utils.data
from torch.utils.data import DataLoader 
from PIL import Image
from matplotlib import pyplot as plt
import seaborn as sns
import math
import argparse
import datetime
import math
from time import time
```

```python
arrk = torch.load("NSE-forecasting-kvals.pt", weights_only=False)
arrtruth = torch.load("NSE-forecasting-truth.pt", weights_only=False)
arrforecst = torch.load("NSE-forecasted-beta-t-EM100.pt", weights_only=False)
```

```python

plt.figure(figsize=(6, 6))
markers = ['o', 's', '^', 'd', 'v']

fig = plt.figure(figsize=(6,6))
plt.plot(arrk, arrtruth, marker=markers[1], markersize=8, markevery=5, linestyle='-', label = 'truth')
plt.plot(arrk, arrforecst, marker=markers[2], markersize=8, markevery=5, linestyle='-', label = 'generated')
plt.legend(fontsize=18, loc='lower left')
plt.grid(alpha=0.3)
plt.xscale('log', base = 2)
plt.yscale('log')
plt.tick_params(axis='both', which='major', labelsize=20)  # Change 14 to your desired size
plt.ylim(1e-5, 1e0)

plt.tight_layout()
plt.savefig(f"NSE-forecasting-beta-t-EM100.pdf", dpi=300)
```
